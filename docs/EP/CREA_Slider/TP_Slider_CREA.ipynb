# Slider CREA

## Présentation


Le rail Edelkrone (figures 1 & 2) est un rail professionnel qui permet un déplacement du chariot sur lequel on fixe une caméra afin de réaliser un travelling motorisé.


Le rail peut être utilisé de deux façons, soit posé sur une table (Figure 1), soit vissé sur un pied photo (Figure 2).

## Procédure de mise en marche (Rail posé sur une table)

1. **Mettre sous tension** le SLIDER, établir la connexion en local de votre ordinateur (ou iPad) au réseau WIFI du SLIDER (SSID du serveur `CREA-SLIDER-XXXX`, mot de passe `sliderCREA`).
2. **Vérifier** que le bouton de blocage central n’empêche pas la course totale du chariot.
3. **Ouvrir** un navigateur internet et **taper** `http://192.168.240.1/SLIDER/`
4. **Cocher** la configuration **`[Rail posé sur une table]`** puis cliquer sur **`[Prise d'origine]`** et
**attester** du déplacement.
5. **Cliquer** sur `PWM80` dans le bandeau supérieur et **actualiser** la page de pilotage et **vérifier** le paramètre
`PWM80`.
6. **Ramener** le banc à la position de repos par la commande : `[Position repos]`

#### Protocole expérimental

1. **Positionner** le mobile à l'origine **`[Prise d'origine]`** ;
