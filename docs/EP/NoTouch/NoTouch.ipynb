# NoTouch

## Présentation

Dans un souci de santé publique, on veut limiter la transmission des bactéries lors du lavage des mains. Le problème vient du fait qu'avec les distributeurs de savon manuels, le contact des mains favorise la transmission des bactéries.
Le système étudié permet la distribution sans contact d'une dose suffisante de savon liquide pour un usage domestique.
Ce distributeur est un nouveau modèle plus compact pour le fabricant. Le cahier des charges spécifie qu'il doit utiliser les mêmes recharges que ceux du modèle précédent.

## Procédure de mise en marche

1. mettre le bouton marche arrêt en position ON en veillant à ne pas placer ses doigts dans le champ de détection du capteur optique ;
2. placer sa main, protégée d’une feuille de papier absorbant, sous l’orifice de distribution du savon ;
3. enlever sa main après distribution du savon.



## Etude du système

> **Mettre** en marche le distributeur en suivant la procédure proposée et **Observer** le mécanisme démonté.
